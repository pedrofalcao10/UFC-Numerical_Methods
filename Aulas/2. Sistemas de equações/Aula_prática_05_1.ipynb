```python
def calcula_solucao_matriz_triangular_inferior(A, B):
    X = np.zeros([len(A), 1], dtype='float')
    i = 0

    while i < len(A):
        z= 0
        soma = 0

        while z < i:
            soma += A[i, z] * X[z]
            z += 1

        X[i] = (B[i] - soma) /  A[i,i]
        i += 1

    return X
```

```python
def gpt_calcula_solucao_matriz_triangular_inferior(A, B):
    n = len(A)
    X = np.zeros((n, 1), dtype=float)

    for i in range(n):
        X[i] = (B[i] - np.dot(A[i, :i], X[:i])) / A[i, i]

    return X
```

```python
def calcula_solucao_matriz_triangular_superior(A, B):
    X = np.zeros([len(A), 1], dtype='float')
    i = len(A) - 1

    while i >= 0:
        z= i + 1
        soma = 0

        while z < len(A):
            soma += A[i, z] * X[z]
            z += 1

        X[i] = (B[i] - soma) /  A[i,i]
        i -= 1

    return X
```

```python
def gpt_calcula_solucao_matriz_triangular_superior(A, B):
    n = len(A)
    X = np.zeros((n, 1), dtype=float)

    for i in range(n - 1, -1, -1):
        X[i] = (B[i] - np.dot(A[i, i+1:], X[i+1:])) / A[i, i]

    return X
```

```python
def eliminacao_gauss(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)

    for i in range(len(A_copia) - 1):
        pivo = A_copia[i, i]

        for j in range(i + 1, len(A_copia)):
            m = -A_copia[j, i] / pivo  # Corrected multiplier calculation
            A_copia[j, :] += A_copia[i, :] * m
            B_copia[j] += B_copia[i] * m

    return A_copia, B_copia
```

```python
A =np.array([[ 1., -3., 2.],
             [-2., 8., -1.],
             [ 4., -6., 5.]])

print(A)

B =np.array([[ 11.],
             [-15.],
             [29.]])

print(B)
```

```python
[A_tri_sup, B_tri_sup] = eliminacao_gauss(A, B)
print(A_tri_sup)
print(B_tri_sup)
```

```python
xr = calcula_solucao_matriz_triangular_superior(A_tri_sup, B_tri_sup)
print(xr)
```

```python
#cálculo do resíduo:
r = B - A@xr
print(r)
```

```python
def calcula_Det(A):
    A_copia = np.copy(A)
    n = len(A_copia)
    det = 1

    for i in range(n - 1):
        pivo = A_copia[i, i]

        for j in range(i + 1, n):
            m = -A_copia[j, i] / pivo  # Corrected multiplier calculation
            A_copia[j, :] += A_copia[i, :] * m

    for i in range(n):
        det *= A_copia[i, i]

    return det
```

```python
A =np.array([[ 1., -3., 2.],
             [-2., 8., -1.],
             [ 4., -6., 5.]])

print(calcula_Det(A))
```
