## *Matriz triangular superior*

```python
def gpt_calcula_solucao_matriz_triangular_superior(A, B):
    n = len(A)
    X = np.zeros((n, 1), dtype=float)

    for i in range(n - 1, -1, -1):
        X[i] = (B[i] - np.dot(A[i, i+1:], X[i+1:])) / A[i, i]

    return X
```

```python
def gpt_calcula_solucao_matriz_triangular_superior(A, B):
    n = len(A)
    X = np.zeros((n, 1), dtype=float)

    for i in range(n - 1, -1, -1):
        X[i] = (B[i] - np.dot(A[i, i+1:], X[i+1:])) / A[i, i]

    return X
```

## *Elimina√ß√£o Gaussiana*

```python
def eliminacao_gauss(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)

    for i in range(len(A_copia) - 1):
        pivo = A_copia[i, i]

        for j in range(i + 1, len(A_copia)):
            m = -A_copia[j, i] / pivo
            A_copia[j, :] += A_copia[i, :] * m
            B_copia[j] += B_copia[i] * m

    return A_copia, B_copia
```

## *Pivoteamento parcial*





```python
def eliminacao_gauss_pivot(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)

    for i in range(len(A_copia) - 1):
        pivo = A_copia[i, i]

        z = i+1
        find = False
        line = i

######################################################## inicio troca das linhas
        while z < len(A_copia):
            if abs(A_copia[z, i]) > abs(pivo):
                pivo = A_copia[z, i]
                line = z
                find = True
            z += 1

        if find:
            Aux = np.copy(A_copia)
            A_copia[i,:] = np.copy(A_copia[line,:])
            A_copia[line,:] = np.copy(Aux[i,:])
            Aux = np.copy(B_copia)
            B_copia[i] = B_copia[line]
            B_copia[line] = Aux[i]
######################################################### final troca das linhas

        for j in range(i + 1, len(A_copia)):
            m = -A_copia[j, i] / pivo
            A_copia[j, :] += A_copia[i, :] * m
            B_copia[j] += B_copia[i] * m

    return A_copia, B_copia
```

```python
def gpt_eliminacao_gauss_pivot(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)
    n = len(A_copia)

    for i in range(n - 1):
        pivot_row = i
        max_pivot = abs(A_copia[i, i])

        # Find the row with the maximum pivot element
        for j in range(i + 1, n):
            current_pivot = abs(A_copia[j, i])
            if current_pivot > max_pivot:
                pivot_row = j
                max_pivot = current_pivot

        # Swap rows if needed
        if pivot_row != i:
            A_copia[[i, pivot_row]] = A_copia[[pivot_row, i]]
            B_copia[i], B_copia[pivot_row] = B_copia[pivot_row], B_copia[i]

        for j in range(i + 1, n):
            m = -A_copia[j, i] / A_copia[i, i]
            A_copia[j, :] += A_copia[i, :] * m
            B_copia[j] += B_copia[i] * m

    return A_copia, B_copia
```

## *Checagem do pivo*


```python
def check_null_pivot(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)

    for i in range(len(A_copia) - 1):
        pivo = A_copia[i, i]

        if pivo == 0:
            z = i+1
            find = False
            line = i

            while z < len(A_copia):
                if abs(A_copia[z, i]) > abs(pivo):
                    pivo = A_copia[z, i]
                    line = z
                    find = True
                z += 1

            if find:
                Aux = np.copy(A_copia)
                A_copia[i,:] = np.copy(A_copia[line,:])
                A_copia[line,:] = np.copy(Aux[i,:])
                Aux = np.copy(B_copia)
                B_copia[i] = B_copia[line]
                B_copia[line] = Aux[i]

    return A_copia, B_copia
```

## *Eliminacao Gauss Jordan*




```python
def eliminacao_gauss_jordan(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)

    for i in range(len(A_copia)):
        pivo = A_copia[i, i]
        A_copia[i,:] /= pivo
        B_copia[i] /= pivo

        for j in range(0, len(A_copia)):
            if j != i:
                m = -A_copia[j, i]
                A_copia[j, :] += A_copia[i, :] * m
                B_copia[j] += B_copia[i] * m

    return A_copia, B_copia
```

## *Inversa via Gauss Jordan*

```python
def inversa_gauss_jordan(A):
    A_copia = np.copy(A)
    B_copia = np.eye(len(A))

    for i in range(len(A_copia)):
        pivo = A_copia[i, i]
        A_copia[i,:] /= pivo
        B_copia[i] /= pivo

        for j in range(0, len(A_copia)):
            if j != i:
                m = -A_copia[j, i]
                A_copia[j, :] += A_copia[i, :] * m
                B_copia[j, :] += B_copia[i, :] * m

    return B_copia
```

## Decomposi√ß√£o LU

```python
def LU_decomposition(A):
    n = A.shape[0]
    U, L = A.copy(), np.eye(n)

    for i in range(n):
        for j in range(i+1, n):
            coef = -U[j, i] / U[i, i]
            L[j, i] = coef
            U[j] += U[i] * coef

    return L, U
```
