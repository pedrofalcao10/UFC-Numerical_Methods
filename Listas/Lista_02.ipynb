## **Fun√ß√µes utilizadas**

Teorema de Bolzano

```python
def bolzano(f, a, b):
  return f(a)*f(b) < 0
```

M√©todo da bissec√ß√£o

```python
def bissection(f, a, b, e, maxiter):
    i = 1
    x = (a + b) / 2

    while i <= maxiter:
        if f(x) == 0:
            return x, f(x)

        if abs(b - a) < e:
            break

        if bolzano(f, a, x):
            b = x
        elif bolzano(f, x, b):
            a = x

        x_ant = x
        x = (a + b) / 2
        erro = abs(x - x_ant) / abs(x)
        print(f'itera√ß√£o:{i}, raiz:{x}, f(x):{f(x)}, erro:{erro}')

        i += 1

    return x, f(x)
```

M√©todo da falsa posi√ß√£o

```python
def false_position(f, a, b, e, max_Iter):
    i = 1
    x = (a * f(b) - b * f(a)) / (f(b) - f(a))
    erro = abs(x - a) / abs(x)

    while i <= max_Iter and erro > e:
        fx = f(x)

        if fx == 0:
            return x, fx

        if bolzano(f, a, x):
            b = x
        elif bolzano(f, x, b):
            a = x

        x_ant = x
        x = (a * f(b) - b * f(a)) / (f(b) - f(a))
        erro = abs(x - x_ant) / abs(x)

        print(f'itera√ß√£o:{i}, raiz:{x}, f(x):{fx}, erro:{erro}')
        i += 1

    return x, f(x)
```

Estimador de itera√ß√µes p/ m√©todo da bissec√ß√£o

```python
def estimate_iterations(a, b, e):
  return math.ceil(((math.log(b - a)) - (math.log(e))) / (math.log(2)))
```

## **Quest√£o 1**

---



Determine a raiz real de $ùëì(ùë•) = 5ùë•^3 ‚àí 5ùë•^2 + 6ùë• ‚àí 2$:

```python
def f1(x):
  return 5*x**3 - 5*x**2 + 6*x - 2
```

### a) Graficamente

```python
x = np.linspace(0, 1, 1000)
y = [f1(i) for i in x]

plt.grid()
plt.plot(x, y)

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.scatter([0.418], [0], color='red', label='Raiz aproximada = 0.418')
plt.legend()

plt.show()
```

### b) Usando o m√©todo da bissec√ß√£o para localizar a raiz. Use as aproxima√ß√µes iniciais $ùë•_ùëô = 0$ e $ùë•_ùë¢ = 1$ e itere at√© que o erro estimado $ùúÄ_ùëé$ fique abaixo de um n√≠vel $ùúÄ_ùë† = 10\%$.


```python
bissection(f1, 0, 1, 0.1, 1e4)
```
