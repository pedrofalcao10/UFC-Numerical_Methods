## *Matriz triangular superior*

```python
def solve_Upper_triang_matrix(A, B):
    n = len(A)
    X = np.zeros((n, 1), dtype=float)

    for i in range(n - 1, -1, -1):
        dot_product = 0.0
        for j in range(i + 1, n):
            dot_product += A[i, j] * X[j]
        X[i] = (B[i] - dot_product) / A[i, i]

    return X
```

## *Matriz triangular inferior*

```python
def solve_Lower_triang_matrix(A, B):
    n = len(A)
    X = np.zeros((n, 1), dtype=float)

    for i in range(n):
        dot_product = 0.0
        for j in range(i):
            dot_product += A[i, j] * X[j]
        X[i] = (B[i] - dot_product) / A[i, i]

    return X
```

## *Pivoteamento parcial*





```python
def eliminacao_gauss_pivot(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)

    for i in range(len(A_copia) - 1):
        pivo = A_copia[i, i]

        z = i + 1
        find = False
        line = i

######################################################## inicio troca das linhas
        while z < len(A_copia):
            if abs(A_copia[z, i]) > abs(pivo):
                pivo = A_copia[z, i]
                line = z
                find = True
            z += 1

        if find:
            Aux = np.copy(A_copia)
            A_copia[i,:] = np.copy(A_copia[line,:])
            A_copia[line,:] = np.copy(Aux[i,:])
            Aux = np.copy(B_copia)
            B_copia[i] = B_copia[line]
            B_copia[line] = Aux[i]
######################################################### final troca das linhas

        for j in range(i + 1, len(A_copia)):
            m = -A_copia[j, i] / pivo
            A_copia[j, :] += A_copia[i, :] * m
            B_copia[j] += B_copia[i] * m

    sol = solve_Upper_triang_matrix(A_copia, B_copia)

    return sol
```

## *Eliminacao Gauss Jordan*




```python
def eliminacao_gauss_jordan(A, B):
    A_copia = np.copy(A)
    B_copia = np.copy(B)

    for i in range(len(A_copia)):
        pivo = A_copia[i, i]
        A_copia[i,:] /= pivo
        B_copia[i] /= pivo

        for j in range(0, len(A_copia)):
            if j != i:
                m = -A_copia[j, i]
                A_copia[j, :] += A_copia[i, :] * m
                B_copia[j] += B_copia[i] * m

    return A_copia, B_copia
```

## *Inversa via Gauss Jordan*

```python
def inversa_gauss_jordan(A):
    A_copia = np.copy(A)
    B_copia = np.eye(len(A))

    for i in range(len(A_copia)):
        pivo = A_copia[i, i]
        A_copia[i,:] /= pivo
        B_copia[i] /= pivo

        for j in range(0, len(A_copia)):
            if j != i:
                m = -A_copia[j, i]
                A_copia[j, :] += A_copia[i, :] * m
                B_copia[j, :] += B_copia[i, :] * m

    return B_copia
```

## Decomposição LU

```python
def LU_decomposition(A):
    n = A.shape[0]
    U, L = A.copy(), np.eye(n)

    for i in range(n):
        for j in range(i+1, n):
            coef = -U[j, i] / U[i, i]
            L[j, i] = -coef
            U[j] += U[i] * coef

    return L, U
```

## Solução do sistema LU

```python
def LU_solution(A, B):
    L, U = LU_decomposition(A)

    Y = solve_Lower_triang_matrix(L,B)
    X = solve_Upper_triang_matrix(U,Y)

    return X
```

##Cálculo do determinante

```python
def det(M):
    if M.shape[0] != M.shape[1]:
        raise ValueError("A matriz não é quadrada.")

    n = M.shape[0]

    if n == 2:
        return M[0, 0] * M[1, 1] - M[0, 1] * M[1, 0]

    determinante = 0
    for j in range(n):
        cofactor = ((-1) ** j) * M[0, j] * det(M[1:, np.arange(n) != j])
        determinante += cofactor

    return determinante
```

## Regra de Cramer

```python
def cramer(M, B):
    n = len(B)
    x = np.zeros((n, 1))
    D = det(M)

    for i in range(n):
        A = M.copy()
        A[:, i] = B[:, 0]
        x[i] = det(A) / D

    return x
```

##Critério de Sassenfeld

```python
def sassenfeld_criteria(A):
    n = len(A)
    B = np.ones(n)

    for j in range(n):
        aux = 0

        for i in range(n):
            if i != j:
                aux += abs(A[j, i]) * B[i]

        B[j] = aux / abs(A[j, j])

    return np.max(B)
```

##Troca-linha Sassenfeld

```python
def troca_linha_sassenfeld2(A, tol=0.1, maxiter=100):
    n, _ = A.shape
    A = A.copy()
    iter_count = 0

    beta_min = sassenfeld_criteria(A)

    while beta_min > tol and iter_count < maxiter:
        i, j = np.random.randint(0, n, size=2)

        A_copy = A.copy()
        A_copy[[i, j]] = A_copy[[j, i]]

        beta = sassenfeld_criteria(A_copy)

        if beta < beta_min:
            beta_min = beta
            A = A_copy

        iter_count += 1

    return sassenfeld_criteria(A), A
```

## Gauss - Seidel

```python
def gauss_seidel(A, B, max_iterations=100, tolerance=1e-8):
    A_copia = A.copy()
    B_copia = B.copy()

    n = len(A)

    X = [np.zeros((n, 1))]

    def relative_difference(x_new, x):
        return np.max(np.abs(x_new - x)) / np.max(np.abs(x_new))

    for i in range(n):
        A_copia[i] = -A[i] / A[i, i]
        A_copia[i, i] = 0
        B_copia[i] = B[i] / A[i, i]

    for _ in range(max_iterations):
        x = X[-1].copy()
        for i in range(n):
            x[i] = A_copia[i] @ x + B_copia[i]
        X.append(x)

        if relative_difference(X[-1], X[-2]) < tolerance:
            break

    return X[-1]
```
