```python
def convert_bin_to_dec(str_bin):
    if "," in str_bin:
        int_part, dec_part = str_bin.split(",")
    else:
        int_part = str_bin
        dec_part = ""

    num_dec = 0

    # Convert the integer part of the binary number
    for index, bit in enumerate(int_part[::-1]):
        num_dec += int(bit) * (2 ** index)

    # Convert the fractional part of the binary number
    for index, bit in enumerate(dec_part):
        num_dec += int(bit) * (2 ** -(index + 1))

    return num_dec

# List of test cases, where each test case is a tuple of (input, expected output)
test_cases = [
    "101101",
    "101,101",
    "0,1101",
    "0,01101",
    "110101011",
    "0,111111101"
]

# Iterate through the test cases and print the results
for idx, input_str in enumerate(test_cases, start=1):
    result = convert_bin_to_dec(input_str)
    print(f"Test case {idx}: Input: {input_str}, Output: {result}")
```

```python
def convert_int_to_bin(num):
    if num == 0:
        return '0'

    binary = ""

    while num > 0:
        binary = str(num % 2) + binary
        num //= 2

    return binary

def convert_real_to_bin(num):
    if num == 0:
        return '0'

    integer_part = int(num)
    decimal_part = num - integer_part

    binary_integer = convert_int_to_bin(integer_part)

    binary_decimal = ''
    while decimal_part > 0 and len(binary_decimal) < 32:
        decimal_part *= 2

        if decimal_part == 1:
            binary_decimal += '1'
            break

        if decimal_part > 1:
            binary_decimal += '1'
            decimal_part -= 1
        else:
            binary_decimal += '0'


    binary_representation = binary_integer + "," + binary_decimal
    return binary_representation

test_cases = [
    37,
    2345,
    0.1217,
    0.125,
    0.1,
    347
]

for i, input_num in enumerate(test_cases, start=1):
    result = convert_real_to_bin(input_num)
    print(f"Test case {i}: Input: {input_num}, Output: {result}")
```
