# 4 – A massa total de uma haste de densidade variável é dada por:

$$
m = ∫_{0}^{L} ρ(x) A_{c}(x) dx
$$

# Em que $m$ é a massa, $ρ(x)$ é a densidade, $A_{c}(x)$ é a área da seção transversal, x é distância ao longo da haste, e L é o comprimento total da haste. Os seguintes dados foram medidos para uma haste de 10m. Determine a massa em quilogramas com a melhor precisão possível.

$$
\begin{array}{|c|c|c|}
\hline
\text{x, m} & \text{0} & \text{2} & \text{3} & \text{4} & \text{6} & \text{8} & \text{10}\\
\hline
\text{ρ, g/cm³} & \text{4.00} & \text{3.95} & \text{3.89} & \text{3.80} & \text{3.60} & \text{3.41} & \text{3.30}\\
\hline
\text{$A_{c}(x)$} & \text{100} & \text{103} & \text{106} & \text{110} & \text{120} & \text{133} & \text{150}\\
\hline
\end{array}
$$

```python
import matplotlib.pyplot as plt

# Given data
x_values = [0, 2, 3, 4, 6, 8, 10]
rho_values = [4.00, 3.95, 3.89, 3.80, 3.60, 3.41, 3.30]
Ac_values = [100, 103, 106, 110, 120, 133, 150]

# Lambda function
rho_Ac_function = lambda x, rho_values, Ac_values: rho_values[x_values.index(x)] * Ac_values[x_values.index(x)]

# Generate points for plotting
x_plot = range(min(x_values), max(x_values) + 1)
y_plot = [rho_Ac_function(x) for x in x_plot]

# Plot the function
plt.plot(x_plot, y_plot, label=r'$\rho(x) \times A_c(x)$')
plt.scatter(x_values, [rho_Ac_function(x) for x in x_values], color='red')  # Plot given data points
plt.xlabel('x')
plt.ylabel(r'$\rho(x) \times A_c(x)$')
plt.title('Function: $\rho(x) \times A_c(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
