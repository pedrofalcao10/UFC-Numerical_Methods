```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fx1(x):
    return math.cos(x)-x**2
```

```python
def flx1(x):
    return -1*math.sin(x)-2*x
```

```python
def met_newton(f, fl, x0, alg_sig, max_iter):
    x = x0 - f(x0)/fl(x0)
    erro = abs(x-x0)/abs(x0)
    delta_x = 10**(-1*(alg_sig+1))
    it = 1

    while abs(x-x0) > delta_x and it < max_iter:
        x0=x
        x = x0 - f(x0)/fl(x0)
        erro = abs(x-x0)/abs(x0)
        it=it+1

    return x

```

```python
[x,erro] = met_newton(fx1,flx1,1,8,100)
```

```python
def fx2(x):
    return math.tan(x)-2*x**2
```

```python
def encontra_intervalo(a0,xmax,h0,f):
    refi = a0
    while h0 > 1e-4:
        b0 = a0+h0
        while b0 < xmax:
            if f(a0)*f(b0) < 0:
                return a0,b0
            a0=b0
            b0=a0+h0
        a0=refi
        h0=h0/10
```

```python
def encontra_varios_intervalos(a0,xmax,h0,f):
    refi = a0
    vet_a0 = []
    vet_b0 = []
    while h0 > 1e-4:
        b0 = a0+h0
        while b0 < xmax:
            if f(a0)*f(b0) < 0:
                vet_a0.append(a0)
                vet_b0.append(b0)
                #return a0,b0
            a0=b0
            b0=a0+h0
        if len(vet_a0) > 0:
            return(vet_a0, vet_b0)

        a0=refi
        h0=h0/10
```

```python
[vxl,vxu] = encontra_varios_intervalos(0,2,0.1,fx2)
print(vxl)
print(vxu)
```

```python
def met_secante(f,x0,x1,alg_sig,max_iter):
    fle = (f(x0)-f(x1))/(x0-x1)
    x = x0 - f(x0)/fle
    print(x)
    erro = abs(x-x0)/abs(x0)
    delta_x = 10**(-1*(alg_sig+1))
    it = 1

    while abs(x-x0) > delta_x and it < max_iter:
        x1 = x0
        x0 = x
        fle = (f(x0)-f(x1))/(x0-x1)
        x = x0 - f(x0)/fle
        print(x)
        erro = abs(x-x0)/abs(x0)
        it=it+1

    return x
```
