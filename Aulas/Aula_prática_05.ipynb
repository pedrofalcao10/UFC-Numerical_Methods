```python
import numpy as np
import matplotlib.pyplot as plt
import math
import time
import copy
```

```python
def calc_pol_conv(vetor,grau,x):
    soma = vetor[0]
    for i in range(1,grau+1):

        soma = soma + vetor[i]*x**i

    return soma
```

```python
def metodo_horner(vetor, grau, x):
    soma = vetor[grau]

    for i in range(0, grau):
        soma = soma*x + vetor[grau-i-1]

    return soma
```

```python
coefic = [1, 1, 0, -4, -3, 2]
```

```python
def estima_qraizes(vet_coef):
  qv = 0
  i=0
  j=1
  while i < len(vet_coef)-1 and j < len(vet_coef):
    if vet_coef[i]*vet_coef[j] < 0:
      qv+=1
      i += 1
      j += 1
    elif vet_coef[i]*vet_coef[j] == 0:
      if vet_coef[j] == 0:
        j += 1
      else:
        i += 1
    else:
      i += 1
      j += 1

  nrp = []
  for i in range(0,qv+1):
    if (qv - i)%2==0:
      nrp.append(i)


  qv = 0
  i=0
  j=1
  for k in range(0,len(vet_coef)):
    vet_coef[k]=vet_coef[k]*((-1)**k)


  while i < len(vet_coef)-1 and j < len(vet_coef):
    if vet_coef[i]*vet_coef[j] < 0:
      qv+=1
      i += 1
      j += 1
    elif vet_coef[i]*vet_coef[j] == 0:
      if vet_coef[j] == 0:
        j += 1
      else:
        i += 1
    else:
      i += 1
      j += 1

  nrn = []
  for i in range(0,qv+1):
    if (qv - i)%2==0:
      nrn.append(i)

  nrc = []
  grau = len(vet_coef)-1
  for i in nrp:
    for j in nrn:
      nrc.append(grau - i - j)

  return nrp,nrn,nrc
```

```python
def signal_changes(vet_coef, i=0, j=1):
    num_changes = 0

    while i < len(vet_coef)-1 and j < len(vet_coef):
        if vet_coef[i] * vet_coef[j] < 0:
            num_changes += 1
            i += 1
            j += 1
        elif vet_coef[i] * vet_coef[j] == 0:
            if vet_coef[j] == 0:
                j += 1
            else:
                i += 1
        else:
            i += 1
            j += 1

    return num_changes
```

```python
def possible_combination_roots(vet_coef):
  num_changes = 0
  i = 0
  j = 1

  num_changes = signal_changes(vet_coef)

  nroots_positive = []
  for i in range(0, num_changes + 1):
      if (num_changes - i) % 2 == 0:
          nroots_positive.append(i)

  num_changes = 0
  i = 0
  j = 1
  for k in range(0,len(vet_coef)):
      vet_coef[k] = vet_coef[k]*((-1)**k)

  num_changes = signal_changes(vet_coef)

  nroots_negative = []
  for i in range(0, num_changes+1):
      if (num_changes - i) % 2 == 0:
          nroots_negative.append(i)

  nroots_complex = []
  grau = len(vet_coef) - 1
  for i in nroots_positive:
      for j in nroots_negative:
          nroots_complex.append(grau - i - j)

  return nroots_positive, nroots_negative, nroots_complex
```
