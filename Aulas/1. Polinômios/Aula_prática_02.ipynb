```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

**MÉTODO DA BISSECÇÃO**

```python
def bolzano(f, a, b):
  return f(a)*f(b) < 0
```

```python
def bissec(a, b, f, e, max_Iter):
  i = 1

  while (i <= max_Iter) and (abs(b-a)/abs(b)) > e:
    x = (a+b) / 2
    print(i, x, f(x))

    if (f(x) == 0) or (abs(b-a)/abs(b) < e):
      return i, x, f(x)

    i += 1

    if bolzano(f, a, x):
      b = x
    elif bolzano(f, x, b):
      a = x
```

```python
def estimate_iterations(a, b, e):
  return math.ceil( ( (math.log(b-a)) - (math.log(e)) ) / (math.log(2)) )
```

**QUESTÃO 1**

```python
def f1(x):
  return math.sqrt(x) - np.cos(x)
```

```python
x1 = np.linspace(0, 1, 1000)
y1 = [f1(i) for i in x1]

plt.grid()
plt.plot(x1, y1, label="Analise de f(x)")
plt.legend()
plt.show()
```

```python
print(estimate_iterations(0, 1, 1e-4))
```

```python
bissec(0, 1, f1, 1e-4, 1e3)
```

**QUESTÃO 2**

```python
def f2(x):
  return 5*math.sin(x**2) - math.exp(x/10)
```

```python
x2 = np.linspace(0, 1, 1000)
y2 = [f2(i) for i in x2]

plt.grid()
plt.plot(x2, y2, label="Analise de f(x)")
plt.show()
```
