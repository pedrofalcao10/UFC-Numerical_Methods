```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return math.sqrt(x)-math.cos(x)
```

```python
def bisseccao(a,b,f,tol,maxiter):
    fa = f(a)
    fb = f(b)
    xr = (a+b)/2
    print(xr)
    fxr = f(xr)
    erro = 1e6
    it = 1

    while it < maxiter and erro > tol:

        if fa*fxr < 0:
            b = xr
            fb = fxr
        elif fb*fxr < 0:
            a = xr
            fa = fxr
        elif fxr == 0:
            return xr

        xra = xr
        xr = (a+b)/2
        print(xr)
        fxr = f(xr)
        erro = abs(xr-xra)/abs(xr)
        it=it+1
    return (xr,it,f(xr))
```

```python
[x_est,nit,fxr] = bisseccao(0,1,f,0.0001,1000)
print(fxr)
```

```python
def estima_iter(a,b,tol):
    x = math.ceil(((math.log(b-a))-(math.log(tol)))/(math.log(2)))
    return x
```

```python
n_it_est = estima_iter(0,1,0.0001)
print(n_it_est)
```

```python
def f(x):
    return 5*math.sin(x**2)-math.exp(x/10)
```

```python
x=np.linspace(0,1,1000)
y=[]
for i in x:
  y.append(f(i))
plt.grid()
plt.plot(x, y, label="Análise de f(x)")
```

```python
def encontra_intervalo(a0,xmax,h0,f):
    refi = a0
    while h0 > 1e-4:
        b0 = a0+h0
        while b0 < xmax:
            if f(a0)*f(b0) < 0:
                return a0,b0
            a0=b0
            b0=a0+h0
        a0=refi
        h0=h0/10
```

```python
[a0,b0] = encontra_intervalo(0,3,0.01,f)
```

```python
print(a0,b0)
```
