```python
v0 =  220 #+-3%
a  = -0.8 #+-4%
v  =  220 #+-1%
i  = (v/v0)**a
print(i)

v0_err = v0*0.03
a_err  =  a*0.04
v_err  =  v*0.01




v0_max = v0 + v0_err
a_max  =  a + a_err
v_max  =  v - v_err
i_max  = (v_max/v0_max)**a_max
print(i_max)




v0_min = v0 - v0_err
a_min  =  a - a_err
v_min  =  v + v_err
i_min  = (v_min/v0_min)**a_min
print(i_min)
```

```python
v0 = 220
delta_v0 = 0.03 * v0  # 3% uncertainty
v = np.linspace(0.99 * v0, 1.01 * v0, 1000)  # Generate a range of v values within a 1% range
delta_v = 0.01 * v  # 1% uncertainty
a = -0.8
delta_a = 0.04  # 4% uncertainty

I = (v / v0) ** a

max_I = ((v - delta_v) / (v0 + delta_v0)) ** (a + delta_a)
min_I = ((v + delta_v) / (v0 - delta_v0)) ** (a - delta_a)

plt.figure(figsize=(10, 6))

# Plot the central value of I
plt.plot(v, I, label='I = (v/v0)^a')

# Plot the maximum and minimum values of I
plt.fill_between(v, min_I, max_I, color='gray', alpha=0.9, label='Uncertainty Range')

plt.xlabel('Voltage (v)')
plt.ylabel('Electric Current (I)')
plt.title('Electric Current vs. Voltage')
plt.legend()
plt.grid(True)
plt.show()
```
