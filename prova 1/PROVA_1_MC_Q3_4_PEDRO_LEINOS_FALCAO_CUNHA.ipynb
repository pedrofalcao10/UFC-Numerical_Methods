```python
#INICIO QUESTﾃグ 4 ---------------------------------------------------------------------------------------------------------

f4 = lambda x : x**4 - 3*x**3 + 5*x**2 - x - 10
```

```python
coef = [-10, -1, 5, -3, 1] #2, -1
grau = 4
```

```python
rho_1 = grau * (abs(coef[0])/abs(coef[1]))
rho_n = (abs(coef[0])/abs(coef[grau]))**(1/grau)
```

```python
print(min(rho_1, rho_n))
```

```python
g4 = lambda x : (3*x**3 - 5*x**2 + x + 10)**(1/4)
```

```python
def fix_point(f, g, x0, tol=0.0001, maxiter=100):
    iter = 1

    while iter < maxiter:
        x_novo = g(x0)
        erro = (x_novo - x0) / x_novo

        print(f'iter={iter}, x_novo:={x_novo}, f(x_novo)={f(x_novo)}, erro={erro}')

        if erro <= tol:
            break

        x0 = x_novo
        iter += 1
    return x_novo
```

```python
fix_point(f4, g4, min(rho_1, rho_n), tol=0.0001, maxiter=100)
```

```python
i_R = 10**(-12)
T = 300
k = 1.380649 * 10**(-23)
q = 1.660217663 * 10**(-19)

vt = k*T/q

req = 1000
V_fonte_a = 30
V_fonte_b = 3
```

```python
def voltagem_diodo(V_fonte, v_diodo, req=1000, i_R=10**(-12), vt=k*T/q):
    return V_fonte - v_diodo - req*i_R*(np.exp(v_diodo/vt) - 1)
```

```python
def bissection(voltagem_diodo, V_fonte, a, b, algsig=3, tol=1e-6, maxiter=100):
    iter = 1
    x_ant = 0

    while iter < maxiter:
        x_atual = (a + b) / 2
        erro = (x_atual - a) / x_atual

        fa = voltagem_diodo(V_fonte, a)
        fb = voltagem_diodo(V_fonte, b)
        fx_atual = voltagem_diodo(V_fonte, x_atual)

        print(f'iter={iter}, voltagem_diodo:={x_atual}, f(x)={voltagem_diodo(V_fonte,x_atual)}, erro={erro}')

        if int((10**algsig)*x_ant)/(10**algsig) == int((10**algsig)*x_atual)/(10**algsig):
            break

        x_ant = x_atual

        if fa*fx_atual < 0:
            b = x_atual
        elif fx_atual*fb < 0:
            a = x_atual

        iter += 1

    return x_atual
```

```python
bissection(voltagem_diodo, V_fonte_a, 0, 1) #ITEM AAAAAAAAAAAAAAAAAAAAAAAAA
```

```python
bissection(voltagem_diodo, V_fonte_b, 0, 1) #ITEM BBBBBBBBBBBBBBBBBBBBBBBBBB
```
