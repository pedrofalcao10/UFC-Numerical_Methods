**FUNCTIONS**

Teorema de Bolzano

```python
def bolzano(f, a, b):
  return f(a)*f(b) < 0
```

Método da falsa posição

```python
def false_position(f, a, b, e, max_Iter):
    i = 1
    erro = 1e6
    x = (a*f(b) - b*f(a)) / (f(b) - f(a))

    while (i <= max_Iter) and erro > e:
      if (f(x) == 0) or erro < e:
        return x, f(x)

      if bolzano(f, a, x):
        b = x
      elif bolzano(f, x, b):
        a = x

      x_ant = x
      x = (a*f(b) - b*f(a)) / (f(b) - f(a))
      erro = abs(x-x_ant)/abs(x)
      print(f'iteração:{i}, raiz:{x}, f(x):{f(x)}, erro:{erro}')

      i += 1
```

Método do Ponto Fixo

```python
def fix_point3(g, x0, a, tol, maxiter):
  x1 = g(a, x0)
  erro = abs(x1 - x0) / abs(x1)
  iter = 1
  verro = []
  vraiz = []

  while erro > tol and iter < maxiter:
    x0 = x1
    x1 = g(a, x0)
    erro = abs(x1 - x0) / abs(x1)
    print(f'iter:{iter}, x1:{x1}, erro:{erro}')

    verro.append(erro)
    vraiz.append(x1)

    iter += 1

  return (x1, verro, vraiz)
```

## **Questão 2**

```python
def f2(y):
  Q = 20
  g = 9.81
  B = 3+y
  Ac = 3*y + (y**2 / 2)

  return 1 - (Q**2 / (g*(Ac**3)))*B
```

## **Questão 4**

```python
def g4a(x):
  return np.log(10/x)
```

```python
def g4b(x):
  return 10*np.exp(-1*x)
```

```python
def fix_point4(g, x0, tol, maxiter):
  x1 = g(x0)
  erro = abs(x1 - x0) / abs(x1)
  iter = 1
  verro = []
  vraiz = []

  while erro > tol and iter < maxiter:
    x0 = x1
    x1 = g(x0)
    erro = abs(x1 - x0) / abs(x1)
    print(f'iter:{iter}, x1:{x1}, erro:{erro}')

    verro.append(erro)
    vraiz.append(x1)

    iter += 1

  return (x1, verro, vraiz)
```

```python
[x1, verro, vraiz] = fix_point4(g4a, 1, 0.001, 100)
```

```python
plt.plot(vraiz)
```

```python
[x1, verro, vraiz] = fix_point4(g4b, 1, 0.001, 100)
```
