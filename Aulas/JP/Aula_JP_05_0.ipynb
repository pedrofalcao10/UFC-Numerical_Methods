```python
import numpy as np
import matplotlib.pyplot as plt
import math
import time
```

```python
grau=5
```

```python
coefic=[1, 1, 0, -4, -3, 2]
```

```python
def calc_pol_conv(vetor,grau,x):
    soma = vetor[0]
    for i in range(1,grau+1):

        soma = soma+vetor[i]*x**i

    return soma
```

```python
def metodo_horner(vetor,grau,x):
    soma = vetor[grau]
    for i in range(0,grau):
        soma = soma*x+vetor[grau-i-1]
    return soma
```

```python
y=calc_pol_conv(coefic,5,7)
print(y)
```

```python
x=np.linspace(0,5,10000)
y1=[]
start = time.time()
for i in x:
    res = calc_pol_conv(coefic,5,i)
    y1.append(res)
end=time.time()
print(end-start)
plt.grid()
plt.plot(x, y1, label="Análise de f(x)")
```

```python
x=np.linspace(0,5,10000)
y2=[]
start = time.time()
for i in x:
    res = metodo_horner(coefic,5,i)
    y2.append(res)
end=time.time()
print(end-start)
plt.grid()
plt.plot(x, y2, label="Análise de f(x)")
```

```python
z = []
for i in range(0,len(y1)):
    z.append(y1[i]-y2[i])

plt.grid()
plt.plot(x, z, label="Análise de f(x)")
```

```python
def estima_qraizes(vet_coef):
  qv = 0
  i=0
  j=1
  while i < len(vet_coef)-1 and j < len(vet_coef):
    if vet_coef[i]*vet_coef[j] < 0:
      qv+=1
      i=i+1
      j=j+1
    elif vet_coef[i]*vet_coef[j] == 0:
      if vet_coef[j] == 0:
        j=j+1
      else:
        i=i+1
    else:
      i=i+1
      j=j+1

  nrp = []
  for i in range(0,qv+1):
    if (qv - i)%2==0:
      nrp.append(i)


  qv = 0
  i=0
  j=1
  for k in range(0,len(vet_coef)):
    vet_coef[k]=vet_coef[k]*((-1)**k)


  while i < len(vet_coef)-1 and j < len(vet_coef):
    if vet_coef[i]*vet_coef[j] < 0:
      qv+=1
      i=i+1
      j=j+1
    elif vet_coef[i]*vet_coef[j] == 0:
      if vet_coef[j] == 0:
        j=j+1
      else:
        i=i+1
    else:
      i=i+1
      j=j+1

  nrn = []
  for i in range(0,qv+1):
    if (qv - i)%2==0:
      nrn.append(i)

  nrc = []
  grau = len(vet_coef)-1
  for i in nrp:
    for j in nrn:
      nrc.append(grau - i - j)

  return nrp,nrn,nrc
```
